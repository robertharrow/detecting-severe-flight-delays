```python
df = pd.read_csv('data/eda/relevant_airports_airlines.csv')
```

```python
df['scheduled-takeoff-utc'] = pd.to_datetime(df['scheduled-takeoff-utc'], utc=True)
df['1-hop-back-scheduled-arrival-utc'] = pd.to_datetime(df['1-hop-back-scheduled-arrival-utc'], utc=True)
df['2-hop-back-scheduled-arrival-utc'] = pd.to_datetime(df['2-hop-back-scheduled-arrival-utc'], utc=True)
df['3-hop-back-scheduled-arrival-utc'] = pd.to_datetime(df['3-hop-back-scheduled-arrival-utc'], utc=True)
df['4-hop-back-scheduled-arrival-utc'] = pd.to_datetime(df['4-hop-back-scheduled-arrival-utc'], utc=True)
df['1-hop-back-scheduled-takeoff-utc'] = pd.to_datetime(df['1-hop-back-scheduled-takeoff-utc'], utc=True)
df['2-hop-back-scheduled-takeoff-utc'] = pd.to_datetime(df['2-hop-back-scheduled-takeoff-utc'], utc=True)
df['3-hop-back-scheduled-takeoff-utc'] = pd.to_datetime(df['3-hop-back-scheduled-takeoff-utc'], utc=True)

df['leg-1-time-diff'] = (df['scheduled-takeoff-utc'] - df['1-hop-back-scheduled-arrival-utc']) / pd.to_timedelta(1, unit='m')
df['leg-2-time-diff'] = (df['1-hop-back-scheduled-takeoff-utc'] - df['2-hop-back-scheduled-arrival-utc']) / pd.to_timedelta(1, unit='m')
df['leg-3-time-diff'] = (df['2-hop-back-scheduled-takeoff-utc'] - df['3-hop-back-scheduled-arrival-utc']) / pd.to_timedelta(1, unit='m')
df['leg-4-time-diff'] = (df['3-hop-back-scheduled-takeoff-utc'] - df['4-hop-back-scheduled-arrival-utc']) / pd.to_timedelta(1, unit='m')

df['origin-4-hops-back-time-diff'] = (df['scheduled-takeoff-utc'] - df['4-hop-back-scheduled-arrival-utc']) / pd.to_timedelta(1, unit='m')
df['origin-3-hops-back-time-diff'] = (df['scheduled-takeoff-utc'] - df['3-hop-back-scheduled-arrival-utc']) / pd.to_timedelta(1, unit='m')
df['origin-2-hops-back-time-diff'] = (df['scheduled-takeoff-utc'] - df['2-hop-back-scheduled-arrival-utc']) / pd.to_timedelta(1, unit='m')
```

```python
df = df.loc[df['leg-1-time-diff'] > 0]
df = df.loc[df['leg-2-time-diff'] > 0]
df = df.loc[df['leg-3-time-diff'] > 0]
df = df.loc[df['leg-4-time-diff'] > 0]
df = df.loc[df['origin-4-hops-back-time-diff'] > 0]
df = df.loc[df['origin-3-hops-back-time-diff'] > 0]
df = df.loc[df['origin-2-hops-back-time-diff'] > 0]
```
